```python
import torch
import numpy as np
import torchvision
import torch.nn as nn
import torchvision.transforms as transforms
```

```python
input_size = 28*28
num_classes = 10
num_epochs=5
batch_size=100
learning_rate=0.01
```

```python
train_dataset = torchvision.datasets.MNIST(root='./MNIST',
                                           train=True,
                                            transform=transforms.ToTensor(),
                                           download=True
                                              )
```

```python
test_dataset = torchvision.datasets.MNIST(root='./MNIST',train=False,
                                         transform=transforms.ToTensor(),
                                         download=True)
```

```python
train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                           shuffle=True,
                                           batch_size=batch_size)
```

```python
test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                         shuffle=True,
                                         batch_size=batch_size)
```

```python
train_dataset[0]
```

```python
model = nn.Linear(input_size,num_classes)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)
```

```python
total_step = len(train_loader)
```

```python
#epoch代表所有数据训练一次
for i in range(num_epochs):
    for j,(images,label) in enumerate(train_loader):
        #因为上面loader的时候batch_size是100，所以这里的images形状是[100,1,28,28]
        # 拉直
        print(images.size())
        images= images.reshape(-1,input_size)
        print(images)
        y_pred = model(images)
        
        
        loss=criterion(y_pred,label)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if(j+1)%10 ==0:
            print('Epoch[{}/{}],step[{}/{}],Loss:{:.4f}'.format(i+1,num_epochs,j+1,total_step,loss.item()))
```

Test

```python
with torch.no_grad():
    correct=0
    total =0
    for images,labels in test_loader:
        images = images.reshape(-1,input_size)
        outputs = model(images)
        # 1的意思可以是axis=1，从行角度看
        _,predicted = torch.max(outputs.data,1)
        print(predicted)
        total +=labels.size(0)
        correct +=(predicted==labels).sum()
        
    print('Accuracy of the model on the 10000 test images: {} %'.format(100 * correct / total))
```
